## Punto 1

```python
x = sym.Symbol('x')
y = sym.Symbol('y')
k = sym.Symbol('k')
```

```python
expr1 = y*x**3 -2*(y**2)*x**2 + x*y**3
sym.integrate(expr1, (y, 0, 2), (x, -1, 1) ) #P1 A
```

```python
f = sym.lambdify((x,y), expr1)
plot3d(expr1, (x, -1, 1),(y, 0, 2), xlabel="x", ylabel="y", size=(8, 8))
```

```python
f = sym.lambdify((x,y), expr1)
h1 = (2-0)/4
h2 = (1 + 1)/4
integral_y = (h1)*(((f(x,0)+f(x,2))/2) + Sum(f(x, 0+k*h1), (k, 1, 3)))

g = sym.lambdify(x, integral_y)
integral_x = (h2)*(((g(-1) + g(1))/2) + Sum(g(-1 + k*h2), (k, 1, 3)).doit())
integral_x # P1 B
```

```python
h1 = (2-0)/2
h2 = (1 + 1)/2
xm = (1 -1)/2
simpson_y = (h1/3)*(f(x, 0) + 4*f(x, h1) + f(x,2))
t = sym.lambdify(x, simpson_y)
simpson_x = (h2/3)*(t(-1) + 4*t(xm) + t(1))
simpson_x # P1 C
```

```python
h1 = (2-0)/3
y_1 = (4+0)/3
y_2 = (2+0)/3

h2 = (1+1)/3
x_1 = (2*(-1)+1)/3
x_2 = (-1 + 2*1)/3

simpson_38_y = ((3*h1)/8)*(f(x, 0) + 3*f(x, y_1) + 3*f(x, y_2) + f(x, 2))
j = sym.lambdify(x, simpson_38_y)
simpson_38_x = ((3*h2)/8)*(j(-1) + 3*j(x_1) + 3*j(x_2) + j(1))
simpson_38_x #P1 D
```

```python
print(-32/9)
```

```python
real = -32/9
print(((real-   integral_x)/real)*100)      # Punto 1 E
print(((real-    simpson_x)/real)*100)
print(((real- simpson_38_x)/real)*100)
```

## Punto 2

```python
expr2 = 2.5*x**5 - 1.2*x**4 + 0.5*x**3 - 0.15*x**2 + 1.3
sym.diff(expr2)
```

```python
sym.plot(expr2, size=(8,8))
```

```python
sym.plot(sym.diff(expr2), size=(8,8))
```

```python
u = sym.lambdify(x, sym.diff(expr2)) 
rel = u(3.3)    # P2 A
rel
```

```python
p = sym.lambdify(x, expr2)

dda = (-p(3.32) +4*p(3.31) -3*p(3.3))/(2*0.01) # P2 B
dda
```

```python
ddb = (3*p(3.3) -4*p(3.29) +p(3.28))/(2*0.01) # P2 C
ddb
```

```python
ddc =(-p(3.32) +8*p(3.31) -8*p(3.29) + p(3.28))/(12*0.01)
ddc
```

```python
print(((rel-   dda)/rel)*100)      # Punto 2 E
print(((rel-   ddb)/rel)*100)
print(((rel-   ddc)/rel)*100)
```
